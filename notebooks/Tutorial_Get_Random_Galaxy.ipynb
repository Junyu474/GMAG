# GMAG Tutorial: Get a Random Galaxy

The initial goal of GMAG (Give Me A Galaxy) is to provide a simple way to get a random galaxy from the SDSS DR17 database. It is not intended to be used in professional environments, but rather as a fun little tool to get a random galaxy to play with, in very few lines of code!

A more practical download tool that query from the SDSS database and download the images is subsequently available in the `gmag.sdss` module. Check out `Tutorial_Download_Images.ipynb` to get more info.

## 0. Importing the Modules

```python
from gmag import sdss
```

## 1. Getting a Random Galaxy

By running the `get_random_galaxy` function, a random galaxy is selected from the SDSS DR17 database. The function returns a `Galaxy` object. The Galaxy image is cutout from a standard SDSS frame, the cutout size is related to the r-band Petrosian radius of the galaxy, also retrieved from SDSS database.

By default, it will print the fetching status. You will get a preview of the galaxy fairly quick, then it will take a few seconds to get the ugriz data.

```python
galaxy = sdss.get_random_galaxy()
```

If you want to disable the fetching status and preview, set `verbose=False` as follows:

```python
galaxy2 = sdss.get_random_galaxy(verbose=False)
```

## 2. Get the Galaxy Information

You can get the galaxy information by calling the `get_info` method. It will print galaxy objid, ra, and dec (more to come).

```python
galaxy.info()
```

## 3. Work with Galaxy Images

The galaxy object stores u,g,r,i,z image data along with a jpg image, all in numpy array format. To access the data:

```python
galaxy.g # g-band image data
```

```python
galaxy.jpg_data # jpg image data
```

Of course, you can also plot the jpg image:

```python
galaxy.show() # show the jpg image
```

or a single band image:

```python
galaxy.show_band(
    'g', # the band to show
    cmap='gray', # optional argument: custom colormap, defaults to 'viridis'
    high_contrast=True, # optional argument: make the galaxy stand out more
    colorbar=True, # optional argument: show the colorbar
)
```

or all the band images:

```python
galaxy.show_all_bands(
    cmap='rainbow', # optional argument: custom colormap, defaults to 'viridis'
    high_contrast=True, # optional argument: make the galaxy stand out more
    colorbar=True, # optional argument: show the colorbar
)
```

That's it! Nothing too complicated here, have fun!
